```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
 
```

```python
df=pd.read_csv("D:/Python Librarys/datasets/multilinearregression.csv")
df
```

```python
reg=linear_model.LinearRegression()

df = df['alan;odasayisi;binayasi;fiyat'].str.split(';', expand=True)
df.columns = ['alan', 'odasayisi', 'binayasi', 'fiyat']

# Convert columns to numeric types
df = df.astype({'alan': 'int', 'odasayisi': 'int', 'binayasi': 'int', 'fiyat': 'int'})

# Fit the regression model
reg.fit(df[['alan', 'odasayisi', 'binayasi']], df['fiyat'])
```

```python
reg.predict([[230,6,0],[500,1,1],[1000,2,0],[1500,3,1],[2000,4,1]])
```

```python
reg.coef_   # b1,b2,b3
```

```python
reg.intercept_# a
```

```python
# y = a + b1*x1 + b2*x2 + b3*x3
# y = a + b1*alan + b2*odasayisi + b3*binayasi

a=reg.intercept_
b1=reg.coef_[0]
b2=reg.coef_[1]
b3=reg.coef_[2]

x1=230
x2=6
x3=3
y=a+b1*x1+b2*x2+b3*x3
print("Tahmin edilen fiyat: ",y)


# Plotting the data
plt.scatter(df['alan'], df['fiyat'], color='blue', label='Alan')
plt.scatter(df['odasayisi'], df['fiyat'], color='red', label='Oda SayÄ±sÄ±')
plt.scatter(df['binayasi'], df['fiyat'], color='green', label='BinayaÅŸÄ±')
plt.xlabel('Ã–zellikler')
plt.ylabel('Fiyat')
plt.title('Ã‡oklu DoÄŸrusal Regresyon')
plt.legend()
plt.show()
```
