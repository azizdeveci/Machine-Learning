```python
import pandas as pd
import matplotlib.pyplot as plt
import os

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Load the dataset
df = pd.read_csv("D:/Python Librarys/datasets/polynomial.csv",sep=";")
df
```

```python
df.columns
```

```python
plt.scatter(df['deneyim'], df['maas'])
plt.xlabel("Deneyim")
plt.xlabel("Deneyim (Yıl)")
plt.ylabel("Maaş (TL)")
plt.title("Deneyim ve Maaş İlişkisi")
plt.savefig("deneyim_maas.png",dpi=300)
plt.show()
```

```python
reg= LinearRegression()
reg.fit(df[['deneyim']],df['maas'])

plt.xlabel("Deneyim(Yıl)")
plt.ylabel("Maaş (TL)")

plt.scatter(df['deneyim'], df['maas'])

x_ekseni=df.deneyim
y_ekseni=reg.predict(df[['deneyim']])

plt.plot(x_ekseni,y_ekseni,color="red",label="Linear Regresyon")
plt.legend()
plt.show()
```

```python
polynomial_regression = PolynomialFeatures(degree=5)

x_polynomial = polynomial_regression.fit_transform(df[['deneyim']])


reg2 = LinearRegression()
reg2.fit(x_polynomial, df['maas']) 
```

```python

y_head= reg2.predict(x_polynomial)
plt.plot(df['deneyim'],y_head,color="red",label="Polinomial Regresyon")
plt.plot(x_ekseni,y_ekseni,color="blue",label="Linear Regresyon")
plt.legend()

plt.scatter(df['deneyim'], df['maas'])
plt.xlabel("Deneyim (Yıl)") 
plt.ylabel("Maaş (TL)")
plt.title("Deneyim ve Maaş İlişkisi")
plt.savefig("deneyim_maas_polynomial.png",dpi=300)
plt.show()  
```
